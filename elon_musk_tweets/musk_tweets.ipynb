## Etapa 1: Carregamento e explora√ß√£o da base de dados

```python
import pandas as pd
```

```python
data_frame = pd.read_csv('./elon_musk_tweets.csv')
```

## Etapa 2: Tratamento dos dados

```python
texts = data_frame['text'].to_list()
```

```python
texts[:3]
```

```python
# Unindo os tweets em uma mesma string
raw_text = ' '.join(texts)
```

```python
chars = sorted(list(set(raw_text)))
```

```python
chars
```

```python
char_to_int = dict((c, i) for i, c in enumerate(chars))
```

```python
char_to_int
```

```python
n_chars = len(raw_text)
n_vocab = len(chars)
print ("Total Caracteres: ", n_chars)
print ("Total Vocabul√°rio: ", n_vocab)
```

## Etapa 3: Split em treino e teste

```python
seq_length = 32
dataX = []
dataY = []

for i in range(0, n_chars - seq_length, 1):
    seq_in = raw_text[i:i + seq_length]
    seq_out = raw_text[i + seq_length]
    dataX.append([char_to_int[char] for char in seq_in])
    dataY.append(char_to_int[seq_out])
n_patterns = len(dataX)
print ("Total de Padr√µes: ", n_patterns)
```

```python
dataX
```

```python
import numpy
```

```python
X = numpy.reshape(dataX, (n_patterns, seq_length, 1))
```

```python
X.shape
```

```python
X
```

```python
# Normaliza√ß√£o
X = X / float(n_vocab)
```

```python
from tensorflow import keras
from keras.utils import np_utils
y = np_utils.to_categorical(dataY)
```

```python
y
```

## Etapa 4: Constru√ß√£o e Treinamento do modelo

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import LSTM
from keras.callbacks import ModelCheckpoint
```

```python
model = Sequential()
model.add(LSTM(256, input_shape=(X.shape[1], X.shape[2]), return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(256))
model.add(Dropout(0.2))
model.add(Dense(y.shape[1], activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam')
```

```python
model.summary()
```

```python
filepath = "weights-improvement-{epoch:02d}-{loss:.4f}.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor = 'loss', verbose = 1, save_best_only = True, mode = 'min')
callbacks_list = [checkpoint]
```

```python
%%time
model.fit(X, y, epochs = 50, batch_size = 140, callbacks = callbacks_list)
```

```python
model.load_weights('weights-improvement-50-1.4474.hdf5')
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam')
```
