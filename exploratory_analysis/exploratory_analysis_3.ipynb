## Division between predictors and responses

```python
x_data_frame = data_frame.iloc[:, 1:4].values
```

```python
y_data_frame = data_frame.iloc[:, 4].values
```

```python
x_data_frame
```

## Handling Categories

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelEncoder = LabelEncoder()
```

```python
x_data_frame[:,2] = labelEncoder.fit_transform(x_data_frame[:,2])
```

```python
x_data_frame[:10, 2]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
columnTransformer = ColumnTransformer(transformers=[('OneHot', OneHotEncoder(), [0, 1])], remainder='passthrough')
x_data_frame = columnTransformer.fit_transform(x_data_frame)
```

```python
x_data_frame
```

```python
x_data_frame[0]
```

```python
x_data_frame.shape
```

## Scaling Values

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_data_frame = scaler.fit_transform(x_data_frame)
x_data_frame
```

```python
x_data_frame[0]
```

## Division between training and test

```python
from sklearn.model_selection import train_test_split
x_training, x_test, y_training, y_test = train_test_split(x_data_frame, y_data_frame, test_size=0.1, random_state=0)
```
